# Machine learning (Random forest)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df_zika_met = pd.read_csv("https://raw.githubusercontent.com/Sirprize123/MasterZika1/main/df_zika_met.csv", sep = ";")
```

```python
df_zika_met.head(10)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
#setando seed
SEED = 42
np.random.seed(SEED)

x = df_zika_met[["Casos confirmados", "PRECIPITAÇÃO TOTAL (mm)",	"TEMPERATURA DO AR - BULBO SECO (°C)",	"UMIDADE RELATIVA DO AR (%)"]]
y = df_zika_met["Data"]

#separando em treino e teste
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.3)

#treino do modelo



modelo = RandomForestClassifier(n_estimators= 200)
modelo.fit(treino_x, treino_y)

#calculando o score
print("resultado da classificação %.2f%%" % (modelo.score(teste_x, teste_y) * 100))
```

Dummy

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(strategy="most_frequent")
dummy.fit(treino_x, treino_y)
acuracia = dummy.score(teste_x, teste_y) * 100

print("A acurácia do dummy stratified foi %.2f%%" % acuracia)
```
